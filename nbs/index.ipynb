# HyperRun

> An open-source "Ask AI" widget for documentation sites — like [RunLLM](https://runllm.com), but self-hosted.

HyperRun lets developers add a semantic code search chat widget to their docs. Built on [ColGrep](https://github.com/lightonai/colgrep) for indexing, [Lisette](https://lisette.answer.ai) for LLM orchestration, and [FastHTML](https://fastht.ml) for the UI.

## Features

- **Semantic code search** — powered by ColGrep's late-interaction retrieval
- **Streaming chat** — SSE-based responses via HTMX
- **Multi-provider** — supports Anthropic, OpenAI, Google and more via LiteLLM
- **BYOK** — users can bring their own API key
- **Chat history** — persistent conversations per session
- **Drop-in widget** — floating "Ask AI" button with DaisyUI-styled chat panel
