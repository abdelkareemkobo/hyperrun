# HyperRun

> An open-source "Ask AI" widget for documentation sites — like [RunLLM](https://runllm.com), but self-hosted.

HyperRun lets developers add a semantic code search chat widget to their docs. Built on [ColGrep](https://github.com/lightonai/colgrep) for indexing, [Lisette](https://lisette.answer.ai) for LLM orchestration, and [FastHTML](https://fastht.ml) for the UI.



## Features


-  **Semantic code search** — powered by ColGrep's late-interaction retrieval
-  **Streaming chat** — SSE-based responses via HTMX
-  **Multi-provider** — supports Anthropic, OpenAI, Google and more via LiteLLM
-  **BYOK** — users bring their own API key
-  **Chat history** — persistent conversations per session
-  **Cost tracking** — per-session token cost display
-  **DaisyUI styled** — floating widget with chat bubbles and markdown rendering
-  **Embeddable** — drop into any static site (Quarto, MkDocs, GitHub Pages) with one snippet


## Quick Start

### 1. Install

```bash
pip install git+https://github.com/abdelkareemkobo/hyperrun.git
```

### 2. Configure

Create a `.env` file:

```
HYPERRUN_REPO_PATH=/path/to/your/codebase
ANTHROPIC_API_KEY=sk-...
```

### 3. Run

```bash
uvicorn hyperrun.ui:app --host 0.0.0.0 --port 8000
```

Visit `http://localhost:8000` to see the widget


## Embed in Your Docs

Add one line to any static site to get the "Ask AI" widget:

```html
<script src="https://YOUR_HYPERRUN_SERVER/embed"></script>
```

### Quarto / nbdev

In `_quarto.yml`:

```yaml
format:
  html:
    include-after-body:
      - text: '<script src="https://YOUR_HYPERRUN_SERVER/embed"></script>'
```

### MkDocs

In `mkdocs.yml`:

```yaml
extra_javascript:
  - https://YOUR_HYPERRUN_SERVER/embed
```

### Any HTML page

```html
<script src="https://YOUR_HYPERRUN_SERVER/embed"></script>
```

Replace `YOUR_HYPERRUN_SERVER` with your deployed HyperRun URL.
